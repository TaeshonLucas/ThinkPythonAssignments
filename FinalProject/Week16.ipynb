```python
# PROGRAM OVERVIEW
# This program allows a user to enter espense records
# Each record contains a category, vendor, amount, and date
# All records are stored in a list of dictionaries
# The user can request totals for:
#   - all expenses
#   - expenses by category
#   - expenses by vendor
#  The program uses a menu-driven loop and runs until the user chooses to exit


# FUNCTION add_expense():
#     ASK user for category
#     ASK user for amount (convert to float)
#     ASK user for date
#
#     CREATE dictionary with these values
#     APPEND dictionary to expenses list
#
#     DISPLAY confirmation message
def add_expense():
    print("\n--- Add New Expense ---")
    category = input("Enter category: ")
    vendor = input("Enter vendor: ")
    # Validate amount input
    while True:
        amount_input = input("Enter amount: ")
        try:
            amount = float(amount_input)
            break
        except ValueError:
            print("Invalid amount. Please enter a number.")

    date = input("Enter date (YYYY-MM-DD): ")

    # expense = {
#     "category": string
#     "vendor": string
#     "amount": float
#     "date": string
# }
    expense = {
      "category": category,
      "vendor": vendor,
      "amount": amount,
      "date": date
    }
    expenses.append(expense)
    print("Expense added successfully!")

# FUNCTION total_all():
def total_all():
#     SET running_total = 0
     running_total = 0
#     FOR each expense in expenses:
     for expense in expenses:
#       ADD expense["amount"] to running_total
       running_total += expense["amount"]
#
#     RETURN running_total
     return running_total


# FUNCTION total_by_category(category_name):
def total_by_category(category_name):
#     SET running_total = 0
    running_total = 0
#     FOR each expense in expenses:
    for expense in expenses:
#       IF expense["category"] == category_name:
#         running_total += expense["amount"]
        if expense["category"].lower() == category_name.lower():
            running_total += expense["amount"]
#     RETURN running_total
    return running_total

# FUNCTION total_by_vendor(vendor_name):
def total_by_vendor(vendor_name):
#     SET running_total = 0
    running_total = 0
#     FOR each expense in expenses:
    for expense in expenses:
#       IF expense["vendor"] == vendor_name:
#         running_total += expense["amount"]
        if expense["vendor"].lower() == vendor_name.lower():
            running_total += expense["amount"]
#     RETURN running_total
    return running_total # Corrected indentation

# A single expense will be represented as a dictionary:
expenses =[]

# All expenses will be stored in a list:
# expenses = []
# expenses.append(expense)
print("---Expense Tracker---")

# LOOP forever:
while True:
#     DISPLAY menu options:
#       1. Add new expense
  print("1. Add new expense")
#       2. Get Totals
  print("2. Get Totals")
#       3. Exit program
  print("3. Exit\n")
#
#     Get user selection
  selection = input("Selection:")
#
#     IF selection == "1":
#       call add_expense()
  if selection == "1":
    add_expense()
#
#     ELSE IF selection =="2":
#       print totaling options, all expenses(A), filter by catecory(C) and filter by vendor(V)
  elif selection == "2":
    print("\n---Totaling Options ---")
    print("A. All Expenses")
    print("C. Expenses by Category")
    print("V. Expenses by Vendor\n")

    # Get selection for totaling options, using a new variable to avoid conflict
    total_selection = input("Selection:")

    if total_selection.upper() == "A":
      total = total_all()
      print(f"\nTotal of all expenses is: ${total:.2f}\n")
    elif total_selection.upper() == "C" :
      category = input("Enter category to total: ")
      total = total_by_category(category)
      print(f"\nTotal for category is: ${total:.2f}\n")
    elif total_selection.upper() == "V" :
      vendor = input("Enter vendor to total: ")
      total = total_by_vendor(vendor)
      print(f"\nTotal for vendor is: ${total:.2f}\n")
    else:
      print("Invalid totaling option\n")
#
#  ELSE IF selection == "3": (exit program)
  elif selection == "3":
    print("Goodbye!")
    break
#  ELSE
#  print invalid option
  else:
    print("Invalid option\n")
```
