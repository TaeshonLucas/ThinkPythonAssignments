```python
import unicodedata

def is_effectively_empty(text):
    if not text:
        return True
    stripped_text = text.strip() # Handle common whitespace
    if not stripped_text:
        return True
    # Check for any remaining characters that are not considered printable/meaningful
    for char in stripped_text:
        # Exclude common invisible characters (e.g., zero-width spaces) and check if printable
        if not unicodedata.category(char).startswith('Z') and char.isprintable():
            return False # Found a meaningful, printable character
    return True # All characters are effectively empty or non-meaningful
```

The `is_effectively_empty` function is a helper to ensure that user input for character names and classes is not just empty, but also doesn't consist solely of whitespace or other non-meaningful characters that `strip()` alone might miss. It checks if a string contains any actual printable, non-spacing characters, providing a more robust definition of 'empty' for user input.

## Implement Calculate Derived Stats Function

### Subtask:
Develop a function that takes a character's base attributes and calculates derived statistics (e.g., Health Points, Attack Power, Defense Rating) based on predefined formulas. This function will add these calculated stats to the character's data.


**Reasoning**:
I will define the `calculate_derived_stats` function as per the instructions, extracting base attributes, calculating derived statistics, and adding them to the character dictionary.



```python
def calculate_derived_stats(character):
    # Safely get attributes with default values
    attributes = character.get('attributes', {})
    strength = attributes.get('strength', 0)
    intelligence = attributes.get('intelligence', 0)
    dexterity = attributes.get('dexterity', 0)

    # Calculate derived statistics
    hp = strength * 10 + intelligence * 2
    attack_power = strength * 3 + dexterity * 1.5
    defense_rating = strength * 2 + dexterity * 1

    # Create a dictionary for derived stats
    derived_stats = {
        "health_points": hp,
        "attack_power": attack_power,
        "defense_rating": defense_rating
    }

    # Add derived stats to the character dictionary
    character['derived_stats'] = derived_stats
    return character

print("Function 'calculate_derived_stats' defined.")
```

**Reasoning**:
I will define the `view_characters` function as per the instructions, checking if the `characters` list is empty, iterating through it, updating derived stats, and then printing all character details in a readable format.



```python
def view_characters():
    if not characters:
        print("No characters have been added yet.")
        return

    print("\n--- Character Roster ---")
    for i, character in enumerate(characters):
        # Ensure derived stats are up-to-date before displaying
        character_with_derived_stats = calculate_derived_stats(character)

        print(f"\nCharacter {i+1}:")
        print(f"  Name: {character_with_derived_stats['name']}")
        print(f"  Class: {character_with_derived_stats['class']}")
        print("  Base Attributes:")
        print(f"    Strength: {character_with_derived_stats['attributes']['strength']}")
        print(f"    Intelligence: {character_with_derived_stats['attributes']['intelligence']}")
        print(f"    Dexterity: {character_with_derived_stats['attributes']['dexterity']}")
        print("  Derived Statistics:")
        print(f"    Health Points: {character_with_derived_stats['derived_stats']['health_points']:.2f}")
        print(f"    Attack Power: {character_with_derived_stats['derived_stats']['attack_power']:.2f}")
        print(f"    Defense Rating: {character_with_derived_stats['derived_stats']['defense_rating']:.2f}")
    print("------------------------")

print("Function 'view_characters' defined.")
```

**Reasoning**:
I will define the `save_characters` function as per the instructions, opening the `FILENAME` in write mode, using `json.dump` to write the `characters` list with indentation, and printing a success message.



```python
def save_characters():
    global characters # Declare characters as global to modify it if needed, though not directly modified here.
    try:
        with open(FILENAME, 'w') as f:
            json.dump(characters, f, indent=4)
        print(f"Characters saved successfully to '{FILENAME}'.")
    except IOError as e:
        print(f"Error saving characters: {e}")

print("Function 'save_characters' defined.")
```

**Reasoning**:
I will define the `load_characters` function as per the instructions, using a `try-except FileNotFoundError` block to handle file loading, and updating the global `characters` list.



```python
def load_characters():
    global characters
    try:
        with open(FILENAME, 'r') as f:
            loaded_data = json.load(f)
            if isinstance(loaded_data, list):
                # Basic validation for each character entry
                validated_characters = []
                for char_data in loaded_data:
                    # Check for essential keys to consider it a valid character dictionary
                    if isinstance(char_data, dict) and \
                       'name' in char_data and \
                       'class' in char_data and \
                       'attributes' in char_data:
                        validated_characters.append(char_data)
                    else:
                        print(f"Warning: Skipping malformed character entry during load: {char_data}")
                characters = validated_characters
                print(f"Characters loaded successfully from '{FILENAME}'. (Some entries might have been skipped due to malformation).")
            else:
                print(f"Error: Data in '{FILENAME}' is not a list. Starting with an empty character list.")
                characters = []
    except FileNotFoundError:
        print("No saved character data found. Starting with an empty character list.")
        characters = []
    except json.JSONDecodeError:
        print(f"Error decoding JSON from '{FILENAME}'. Starting with an empty character list.")
        characters = []

print("Function 'load_characters' defined.")
```

## Integrate Save and Load Functionality

### Subtask:
Add functions to save the current list of characters to a JSON file and to load characters from that file when the program starts. This ensures data persistence across sessions.


**Reasoning**:
I will define the `main_menu` function that contains the main program loop, displays the menu options, handles user input to call the appropriate functions (`add_character`, `view_characters`, `save_characters`, `load_characters`), and includes an option to exit the program. Finally, I will call `main_menu()` to start the application.



```python
def main_menu():
    print("\nWelcome to the RPG Character Planner!")
    load_characters() # Load characters at the start of the program

    while True:
        print("\n--- Main Menu ---")
        print("1. Add New Character")
        print("2. View Characters")
        print("3. Save Characters")
        print("4. Load Characters")
        print("5. Exit")
        print("-----------------")

        try:
            choice = input("Enter your choice (1-5): ")
        except KeyboardInterrupt:
            print("\nExiting RPG Character Planner. Goodbye!")
            break # Exit the main loop on KeyboardInterrupt

        if choice == '1':
            add_character()
        elif choice == '2':
            view_characters()
        elif choice == '3':
            save_characters()
        elif choice == '4':
            load_characters()
        elif choice == '5':
            print("Exiting RPG Character Planner. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

# Start the program
if __name__ == "__main__":
    main_menu()

print("Function 'main_menu' defined and program started.")
```
