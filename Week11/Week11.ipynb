```python
# rectangle.py
# IMPORT STATEMENT

# FUNCTION DEFINITION
# Define calculate_rectangle(length, width):
def calculate_rectangle(length, width):
#   set area = length * width
  area = length * width
#   set perimeter = 2 * (length + width)
  perimeter = 2 * (length + width)
#   return area and perimeter
  return area, perimeter

# PROGRAM LOGIC
#   Display "Enter the rectangle length: "
#   get user input,convert to float, store in length
length = float(input("Enter the rectangle length: "))
#   Display "Enter the rectangle width"
#   get user input,convert to float, store in length
width = float(input("Enter the rectangle width: "))
#   Call calculate_rectangle(length, width)
#   store returned values into my_area and my_perimeter
my_area, my_perimeter = calculate_rectangle(length, width)
#   Display results
print("\n--- Rectangle Result ---")
print(f"Area: {my_area}")
print(f"Perimeter: {my_perimeter}")
```
