```python
# NAME: gradebook.py
# IMPORT STATEMENTS

# FUNCTION DEFINITION
# add_students(gradebook)
def add_students(gradebook):
#   use while loop
    while True:
#   ask student for name until user types "done"
        name = input("Enter student name (or 'done'): ")
        if name == "done": # Corrected indentation for if and break
            break
#   if not done, ask for grade
# grade = input("Enter student name :")`
        grade = int(input("Enter student grade:"))
        gradebook[name] = grade
# calculate_average(gradebook)
def calculate_average(gradebook):
  if len(gradebook) == 0:
   return None
#   get all grade values using gradebook.values()
  grades = gradebook.values()
#   compute average
  total = sum(grades)
  average = total / len(grades)
#   return the average
  return average

# PROGRAM LOGIC
#   create empty gradebook dictionary
gradebook = {}
#   call add_students to fill dictionary
add_students(gradebook)
#   call calculate_average to get class average
class_average = calculate_average(gradebook)
#   print each student and their grade
print("\n---Class Report---")
print("\nTesting gradebook contents:")
#   print formatted class average
print(f"Class Average: {class_average: .2f}")
print(gradebook)
```
