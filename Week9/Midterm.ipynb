```python
# NAME: rentalcar.py
# IMPORT STATEMENTS

#FUNCTION DEFINITIONS

# calculate_miles(start, end)
def calculate_miles(start, end):
  return end - start

# calculate_bill(code, days, miles)
def calculate_bill(code, days, miles):
  daily_rate = 0
  mileage_charge = 0

  if code == "B":
      daily_rate = 40
      mileage_charge = miles * 0.25
  elif code == "S":
      daily_rate = 60
      free_miles = 100 * days
      charged_miles_for_calculation = max(0, miles - free_miles)
      mileage_charge = charged_miles_for_calculation * 0.25
  elif code == "L":
      daily_rate = 85
      mileage_charge = miles * 0.30

  total_bill = daily_rate * days + mileage_charge
  return total_bill

# get_classification_name(code)
def get_classification_name(code):
  if code == 'B':
    return 'Budget'
  if code == 'S':
    return 'Standard'
  if code == 'L':
    return 'Luxury'


# PROGRAM LOGIC
# ask user for classification code
code = input("Enter classification code: ").upper()
# while code note valid:
while code not in ["B", "S", "L"]:
#     ask again
 code = input("Invalid code. Enter classification code: ").upper()

# ask for start odemeter
start = int(input("Start Odemeter: "))
# ask for end odemeter
end = int(input("End Odemeter: "))
# ask for days rented
days = int(input("Days Rented: "))

# compute miles using calculate_miles()
miles = calculate_miles(start, end)
# compute bill using calculate_bill()
bill = calculate_bill(code, days, miles)

# Calculate charged_miles for printing outside the function
if code == 'S':
  free_miles = 100 * days
  charged_miles = max(0, miles - free_miles)
else:
  # For B and L, all miles are charged (or effectively, all miles contribute to mileage_charge)
  charged_miles = miles

# print rental summary report
print("\n--- RENTAL SUMMARY ---")
print("Classification:", get_classification_name(code))
print("Total Miles:", miles)
print("Charged Miles:", charged_miles)
print(f"Total Bill: ${bill}")
```
