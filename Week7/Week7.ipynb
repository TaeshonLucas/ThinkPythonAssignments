```python
# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
# 1. Define verify_pin():
def verify_pin():
  pin = ""
  while pin != "1234":
    pin = input("Enter your PIN:")
    if pin == "1234":
      print("Correct PIN.")
      return True
    else:
      print("Incorrect PIN. Try again.")

#    - Create a loop that asks for a PIN until it equals "1234"
#    - Return True if the PIN is correct

# 2. Define withdraw(balance, amount):
def withdraw(balance, amount):
#    - Check if enough money exists
  if amount > balance:
    print("Insufficient funds.")
    return balance
  else:
    return balance - amount
    print(f"${amount} withdrawn.")
#    - Return the new(or old) balance
  return balance
# PROGRAM LOGIC
# 1. Set current_balance to 500
current_balance =500
# 2. Call verify_pin()
verify_pin()
# 3. Get withdrawal amount from user input
amount = float(input("Enter withdrawal amount:"))
# 4. Update current_balance by calling withdraw()
current_balance = withdraw(current_balance, amount)
# 5. Print the final balance
print("Withdrawal successful.")
print("Final balance:" "$" + str(current_balance))
```
