```python
# IMPORT STATEMENT
# (None)

# FUNCTION DEFINITION
# Define function is_palidrome that accepts one word
def is_palindrome(word):
# Convert word to lower case so capitalization doesn't matter
    word = word.lower()
# Reverse word using string slicing
    reversed_word = word[::-1]
# Return True if they match, otherwise return False
    return word == reversed_word

# PROGRAM LOGIC
# Prompt the user to enter a word. If true, loop
user_word = input("Enter a word:")
# Call the is_palindrome function using the user's input
result = is_palindrome(user_word)
# If the function returned True, print a message saying the word Is a palindrome
if result:
   print(f"Yes! {user_word} is a palindrome.")
# Otherwise, print a message saying the word is Not a palindrome
else:
   print(f"No. {user_word} is not a palindrome.")
```
