# Personal Task Manager

```python
tasks = []
COMPLETED_TASKS_FILE = "completed_tasks.txt" # Define file name

def add_task(task_name, priority):
    tasks.append({"name": task_name, "priority": priority})
    print(f"Task '{task_name}' with priority {priority} added.")

def view_tasks():
    if not tasks:
        print("No tasks to display.")
        return [] # Return empty list if no tasks

    # Sort tasks by priority (1: High, 2: Medium, 3: Low)
    sorted_tasks = sorted(tasks, key=lambda x: x['priority'])

    print("\n--- Your Tasks ---")
    for i, task in enumerate(sorted_tasks):
        priority_map = {1: "High", 2: "Medium", 3: "Low"}
        print(f"{i+1}. Name: {task['name']}, Priority: {priority_map.get(task['priority'], 'Unknown')}")
    print("------------------\n")
    return sorted_tasks # Return the sorted list

def complete_task():
    sorted_tasks = view_tasks()
    if not sorted_tasks:
        return

    while True:
        try:
            task_num_to_complete = int(input("Enter the number of the task to complete: "))
            if 1 <= task_num_to_complete <= len(sorted_tasks):
                completed_task_obj = sorted_tasks[task_num_to_complete - 1]

                # Find and remove the exact task object from the original 'tasks' list
                # This ensures we remove the correct dictionary object by identity
                for i, task in enumerate(tasks):
                    if task is completed_task_obj:
                        tasks.pop(i)
                        break

                task_name = completed_task_obj['name']

                with open(COMPLETED_TASKS_FILE, "a") as f: # "a" for append, creates file if not exists
                    f.write(f"{task_name}\n")
                print(f"Task '{task_name}' marked as complete and saved to {COMPLETED_TASKS_FILE}.")
                break
            else:
                print("Invalid task number. Please enter a valid number from the list.")
        except ValueError:
            print("Invalid input. Please enter a number.")
        except IndexError:
            print("Task not found.") # Should not happen with prior validation

def what_is_next():
    priority_1_tasks = [task for task in tasks if task['priority'] == 1]
    if priority_1_tasks:
        # Since tasks are added to the list chronologically, the first one found is the oldest.
        # If we wanted the oldest from sorted tasks, we'd need a more complex sort key or store timestamps.
        # For now, assuming insertion order implicitly means oldest for un-removed tasks.
        oldest_p1_task = priority_1_tasks[0]
        print(f"\nYour next high-priority task is: '{oldest_p1_task['name']}' (Priority: High)")
    else:
        print("\nGreat job! No Priority 1 tasks are currently pending. You're doing awesome!")

while True:
    print("\nPersonal Task Manager Menu:")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Complete Task")
    print("4. What is next?")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        task_name = input("Enter task name: ")
        while True:
            try:
                priority = int(input("Enter priority (1 for High, 2 for Medium, 3 for Low): "))
                if priority in [1, 2, 3]:
                    add_task(task_name, priority)
                    break
                else:
                    print("Invalid priority. Please enter 1, 2, or 3.")
            except ValueError:
                print("Invalid input. Please enter a number.")
    elif choice == '2':
        view_tasks()
    elif choice == '3':
        complete_task()
    elif choice == '4':
        what_is_next()
    elif choice == '5':
        print("Exiting Task Manager. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```
