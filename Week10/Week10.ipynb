```python
# NAME: roster.py
# IMPORT STATEMENTS


# FUNCTION DEFINITIONS

# add_players(roster)
def add_players(roster):
#   loop while True
  while True:
#   ask user for a name
    name = input("Enter your name(or 'quit' to stop): ")
#   if name == "quit": break
    if name.lower() == "quit":
      break
#   otherwise append the name to the roster list
    else:
      roster.append(name)

# is_on_roster(roster, name)
def is_on_roster(roster, name):
#   loop through each player in roster
  for player in roster:
    if player.lower() == name.lower():
#   if player matches the name, return True
      return True
#   after loop ends, return False
  return False


# PROGRAM LOGIC
#   create empty list called roster
roster = []
#   call add_players(roster)
add_players(roster)
#   ask user for a name to search for
search_name = input("\nWho would you like to search for? ")
#   call is_on_roster and store the result
if is_on_roster(roster, search_name):
  print("Result: Player found!")
else:
  print("Result: Player not found.")
#   print message depending on True/False
print("Current_roster: ", roster)
```
