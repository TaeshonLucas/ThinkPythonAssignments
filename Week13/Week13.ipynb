```python
# NAME: journal.py
# IMPORT STATEMENTS
import os

# FUNCTION DEFINITION

# write_entry(filename, entry)
def write_entry(filename, entry):
#   open the file in append mode ('a')
  file = open(filename, 'a')
#   write the entry followed bya newline character("\n")
  file.write(entry + "\n")
#   close the file
  file.close()

# read_journal(filename)
def read_journal(filename):
  if not os.path.exists(filename):
    print("No entries found yet. The journal file does not exist.")
    return

  try:
#   open the file in read mode ('r')
    file = open(filename, 'r')
#   read the entire contents
    contents = file.read()
#   close the file
    file.close()
#   print the contents or message if empty
    if not contents.strip(): # Check if content is empty or just whitespace
      print("No entries found yet. The journal file is empty.")
    else:
      print(contents)
  except IOError as e: # Catch general I/O errors during file operations
    print(f"An error occurred while reading the journal: {e}")

# PROGRAM LOGIC
filename = "journal.txt"
#   while True:
#     ask user: Write, Read, or quit
while True:
  choice = input("\nWrite, Read, or quit? ")
#     if W:
#       ask for journal entry text
  if choice == "w":
#       call write_entry
    text = input("Write your entry: ")
    write_entry(filename, text)
    print("Entry saved.")
#     if R:
#       call read_journal
  elif choice == "r":
    print("Reading journal entries...")
    read_journal(filename)
#     if quit:
#       break loop
  elif choice == "q":
    print("Goodbye.")
    break
  else:
    print("Invalid choice.")
```
